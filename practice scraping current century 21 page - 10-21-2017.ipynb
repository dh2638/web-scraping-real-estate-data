```python
import requests 
from bs4 import BeautifulSoup 
```

```python
r = requests.get("https://www.century21.com/real-estate/new-york/LSNY/?pr=%7C200000&be=4&sn=5&sk=Y&pt=2") 
c = r.content
```

```python
c
```

```python
soup = BeautifulSoup(c, "html.parser")
```

```python
soup
```

```python
print(soup.prettify())
```

```python
all = soup.find_all("div",{"class":"property-card-primary-info"})
```
